# Training Classifier dengan `Decision Tree`

Pada kali ini saya akan melakukan *training classifier* dengan beberapa model machine learning yang telah dipelajari yang pada kali ini saya akan memulai klasifikasi dengan **Desicion Tree**.

# Classification Task dengan Decision Tree

# Load Breast Cancer Dataset

```python
# Code 01

from sklearn.datasets import load_breast_cancer

X, y = load_breast_cancer(return_X_y=True)

print(f'Dimensi Feature: {X.shape}')
print(f'Class: {set(y)}')
```

**Konsep berpikir code 01 :**

- Pertama-tama kita akan import terlebih dahulu modulnya dengan memanggil "from sklearn.datasets import load_breast_cancer".
- Lalu berikutnya, kita akan panggil method "load_breast_cancer" dengan menyertakan parameter (return_X_y=True) dengan konsekuensi kita harus menyertakan variable X dan Y. Dimana variabel X akan digunakan untuk menampung sekumpulan nilai feature dan variabel y akan digunakan untuk menampung sekumpulan nilai target labelnya.
- Selanjutnya kita akan cobba tampilkan pada layar dimensi dari feature nya dengan memanggil method "X_shape" dan juga kita akan cetak pada layar class apa saja yang terdapat dalam target labelnya.

Bisa kita lihat pada output code 01 bahwa dimensi data dari feature nya adalah (569, 30) artinya disini terdapat 569 baris dan 30 kolom. Lalu terkait dengan class pada target labelnya, terdapat 2 buah class yaitu {0, 1}.

Selanjutnya kita akan bagi dataset tersebut kedalam training dan testing set (perhatikan pada code 02).

```python
# Code 02

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=0)
```

**Konsep berpikir code 02 :**

- Disini kita akan mengimport modul dengan memanggil method "from sklearn.model_selection import train_test_split".
- Untuk selanjutnya kita akan panggi train_test_split nya, lalu kita sertakan X dan y sebagai parameternya.
- Lalu terkait parameter test_size nya kita set sebagai 0.2 yang artinya disini testing set nya akan menempati 20% dari total keseluruhan dataset yang kita miliki.
- Lalu berikutnya untuk random_state kita beri nilai 0.
- Pemanggilan fungsi train_test_split ini akan menghasilkan 4 kumpulan data, dimana keempatnya perlu kita tampung kedalam 4 buah variabel yaitu X_train, X_test, y_train, dan y_test.

# 1. Classification dengan `DecisionTreeClassifier`

Disini kita akan menerapkan *Desicion Tree* untuk melakukan klasifikasi **Breast Cancer**.

```python
# Code 02

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=4)

model.fit(X_train, y_train)
```

**Konsep berpikir code 02 :**

- Pertama-tama kita akan import terlebih dahulu modulnya dengan memanggil method "from sklearn.tree import DecisionTreeClassifier".
- Lalu berikutnya kita akan bentuk objek dari class Desicion Tree Classifier dengan menyertakan parameter (max_depth) yang kita set sebafai 4 yang berarti Desicion Tree Classifier yang kita hasilkan itu memiliki kedalaman maksimum 4 layer. Berikut objek dari Desicion Tree Classifier tersebut akan kita tampung kedalam variabel model.
- Untuk berikutnya akan kita training dengan memanggil method "model.fit" yang dimana proses training ini akan melibatkan (X_train, y_train) sebagai training setnya.

# Visualisasi Model

Disini kita akan melihat visulisasi model dari **Breast Cancer Dataset**. Berikut adalah caranya.

# Evaluasi Model 

Selanjutnya kita akan melakukan evaluasi performa dari  **Desicion Tree** model yang baru saja kita training tadi dengan memanfaatkan *Classification Report*.

```python
# Code 04

from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

**Konsep berpikir code 04 :**

- Disini pertama-tama kita akan import terlebih dahulu modulnya dengan memanggil method "from sklearn.metrics import classification_report".
- Lalu berikutnya kita akan melakukan prediksi dengan memanggil method "model.predict" dengan memanfaatkan testing_set kita (X_test), yang hasil prediksinya akan ditampung kedalam variabel y_pred.
- Untuk selanjutnya akan kita evaluasi dengan menggunakan classification_report dengan menyertakan parameter y_test dan y_pred.
- Pada output code 04, kita bisa melihat kolom nilai precission, recall, f1-score, support, serta kolom nilai accuracy nya sebagai peforma dari **Desicion Tree** model yang telah kita gunakan.

```python
from sklearn.metrics import classification_report
import numpy as np
import pandas as pd

from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV


X, y = make_classification(n_samples=1000, n_features=30,
                           n_informative=12,
                           n_clusters_per_class=1, n_classes=10,
                           class_sep=2.0, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y)


clf = LogisticRegression(max_iter=1000, random_state=42).fit(X_train, y_train)



df = pd.DataFrame(classification_report(clf.predict(X_test), 
                                        y_test, digits=2,
                                        output_dict=True)).T

df['support'] = df.support.apply(int)

df.style.background_gradient(cmap='viridis',
                             subset=pd.IndexSlice['0':'1', :'f1-score'])
```
